```python

tweet_params = []
with open('tweet_json.txt', encoding='utf-8') as  file:
   
    
    #for line in glob.glob('tweet_json.txt'):
    line = file.readlines()[:-1]
    for tweet_line in line:
    
        
        TWEET_LINE = json.loads(tweet_line )
       # TWEET_LINE = JSONDecoder(jstr)
        tweet_ID = TWEET_LINE['id']
        #TEXT = json.loads(tweet_line)
        text = TWEET_LINE['text']
        #RETWEET_COUNT = json.loads(tweet_line)
        retweet_count = TWEET_LINE['retweet_count']
        favorite_count = TWEET_LINE['favorite_count']
        truncated = TWEET_LINE['truncated']
        entities = TWEET_LINE['entities']
        #extended_entities = TWEET_LINE['extended_entities']
        source = TWEET_LINE['source']
        in_reply_to_status_id = TWEET_LINE['in_reply_to_status_id']
        in_reply_to_status_id_str = TWEET_LINE['in_reply_to_status_id_str']
        in_reply_to_user_id = TWEET_LINE['in_reply_to_user_id']
        in_reply_to_user_id_str = TWEET_LINE['in_reply_to_user_id_str']
        in_reply_to_screen_name = TWEET_LINE['in_reply_to_screen_name']
        user = TWEET_LINE['user']
        geo = TWEET_LINE['geo']
        contributors = TWEET_LINE['contributors']
        is_quote_status = TWEET_LINE['is_quote_status']
        favorited = TWEET_LINE['favorited']
        retweeted = TWEET_LINE['retweeted']
       # possibly_sensitive = TWEET_LINE['possibly_sensitive']
       # possibly_sensitive_appealable = TWEET_LINE['possibly_sensitive_appealable']
        lang = TWEET_LINE['lang']
        tweet_params.append({'tweet_ID': tweet_ID,
                'text':text,
               'retweet_count':retweet_count,
               'favorite_count': favorite_count,
               'truncated': truncated,
               'entities': entities,
               #'extended_entities': entended_entities,
               'source': source,
               'in_reply_to_status_id':in_reply_to_status_id ,
               'in_reply_to_status_id_str': in_reply_to_status_id_str,
               'in_reply_to_user_id': in_reply_to_user_id,
               'in_reply_to_user_id_str': in_reply_to_user_id_str,
               'in_reply_to_screen_name':in_reply_to_screen_name,
               'user': user,
               'geo': geo,
               'contributors': contributors,
               'is_quote_status': is_quote_status,
               'favorited': favorited,
               'retweeted': retweeted,
               #'possibly_sensitive': possibly_sensitive,
               #'possibly_sensitive_appealable':possibly_sensitive_appealable,
               'lang': lang
                })
```

```python
import pandas as pd
#pd.set_option('display.max_rows', 1000)
df = pd.DataFrame(tweet_params, columns = ['tweet_ID', 'text', 'retweet_count', \
                                           'favorite_count', 'truncated', 'entities',\
                                           'source', 'in_reply_to_status_id', \
                                           'in_reply_to_status_id_str',\
                                           'in_reply_to_user_id','in_reply_to_user_id_str',\
                                           'in_reply_to_screen_name', 'user','contributors',\
                                           'is_quote_status','favorited','retweeted'])
df
```

```python
denominator = []
numerator = []
name = []
for text in df2:
    slash = text.find('/')
   # name_start = text.find(' ', 6)
    name_start = text.find('This is ')
    name_end = text.find('.')
    nam = text[name_start+8 : name_end]
    den = text[slash+1 : slash+3]
    num = text[slash-2 : slash]
    denominator.append(den)
    numerator.append(num)
    name.append(nam)
```

```python
#denominator
#numerator
df['numerator'] = numerator
df['denominator'] = denominator
df['name'] = name
```
